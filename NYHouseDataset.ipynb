```python
import pandas as pd

# Example criteria for a good investment property
min_bedrooms = 3
min_bathrooms = 2
min_sqft = 1500
max_price = 500000  # Adjust this based on your budget

# Filtering the DataFrame
good_investment_properties = df[
    (df['BEDS'] >= min_bedrooms) &
    (df['BATH'] >= min_bathrooms) &
    (df['PROPERTYSQFT'] >= min_sqft) &
    (df['PRICE'] <= max_price)
]

# Displaying the result
print(good_investment_properties)
```

```python
# Display names in the 'LOCALITY' column
unique_localities = df['LOCALITY'].unique()
print("Unique Localities:")
print(unique_localities)
# Display names in the 'SUBLOCALITY' column
unique_sublocalities = df['SUBLOCALITY'].unique()
print("\nUnique Sublocalities:")
print(unique_sublocalities)
```

```python
# Classify for sublocalities to major boroughs
borough_mapping = {'Manhattan': ['Manhattan', 'New York County', 'New York'],
    'Queens': ['Queens', 'Queens County', 'Jackson Heights', 'Rego Park', 'Flushing'],
    'Brooklyn': ['Brooklyn', 'Kings County', 'Coney Island', 'Brooklyn Heights', 'Fort Hamilton', 'Dumbo', 'Snyder Avenue'],
    'Staten Island': ['Staten Island', 'Richmond County'],
    'The Bronx': ['The Bronx', 'East Bronx', 'Bronx County', 'Riverdale'],}

# Map sublocalities to boroughs
df['BOROUGH'] = df['SUBLOCALITY'].apply(lambda subloc: next((borough for borough, sublocs in borough_mapping.items() if subloc in sublocs), None))

unique_boroughs = df['BOROUGH'].unique()
print("\nUnique Boroughs:")
print(unique_boroughs)
```

```python
# Group by 'BOROUGH' and calculate the average housing price for each borough
average_housing_price_per_borough = df.groupby('BOROUGH')['PRICE'].mean()
print("\nAverage Housing Price per Borough:")
print(average_housing_price_per_borough)
```

```python
# Group by 'BOROUGH' and calculate the average square footage for each borough
average_square_footage_per_borough = df.groupby('BOROUGH')['PROPERTYSQFT'].mean()
print("\nAverage Square Footage per Borough:")
print(average_square_footage_per_borough)
```

```python
# Group by 'BOROUGH' and 'Type' to show the count of each property type for each borough.
property_type_counts = df.groupby(['BOROUGH', 'TYPE']).size().reset_index(name='Count')
print("\nCount of Each Property Type for Each Borough:")
print(property_type_counts)
```

```python
# Create a bar plot for the count of each property type per borough
plt.figure(figsize=(14, 8))
sns.countplot(x='BOROUGH', hue='TYPE', data=df)
plt.title('Count of Each Property Type for Each Borough')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.legend(title='Property Type', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

```python
# Calculate the "Square Footage per Price" ratio
df['SQFT_PER_PRICE'] = df['PROPERTYSQFT'] / df['PRICE']

# Group by 'BOROUGH' and calculate the average of "Square Footage per Price"
average_sqft_per_price_per_borough = df.groupby('BOROUGH')['SQFT_PER_PRICE'].mean()
print("\nAverage Square Footage per Price per Borough:")
print(average_sqft_per_price_per_borough)
```

```python
# Create a bar plot for average square footage per price per borough
average_sqft_per_price_per_borough = average_sqft_per_price_per_borough.sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=average_sqft_per_price_per_borough.index, y=average_sqft_per_price_per_borough.values)
plt.title('Comparison of Average Property Value In Square Feet across Boroughs')
plt.xlabel('Borough')
plt.ylabel('Average Square Footage per Price')
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# Group by 'BOROUGH' and print the number of beds and baths
beds_baths_counts_per_borough = df.groupby('BOROUGH')[['BEDS', 'BATH']].count()
print("\nCounts of Beds and Baths per Borough:")
print(beds_baths_counts_per_borough)
```

```python
# Group by 'BOROUGH' and print the number of beds and baths
beds_counts_per_borough = df.groupby('BOROUGH')[['BEDS']].count()
print("\nCounts of Beds per Borough:")
print(beds_counts_per_borough)
```

```python
# Group by 'BOROUGH' and calculate the average housing price and counts of beds
average_housing_price_per_borough = df.groupby('BOROUGH')['PRICE'].mean()
beds_counts_per_borough = df.groupby('BOROUGH')['BEDS'].count()
```

```python
# Calculate average housing price per bed per borough
average_price_per_bed_per_borough = average_housing_price_per_borough / beds_counts_per_borough
print("\nAverage Housing Price per Bed per Borough:")
print(average_price_per_bed_per_borough)
```
